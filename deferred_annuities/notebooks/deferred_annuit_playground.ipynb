# How to Use the Annuity Valuation Plotter

This tool provides a visual and computational way to solve deferred annuity problems. It automatically calculates the exact values (PV, FV, Current Value) and generates an interactive timeline graph.

## Function Parameters

To generate a plot, import and call the `plot_annuity_valuation` function with the following parameters:

| Parameter | Type | Default | Description |
| :--- | :--- | :--- | :--- |
| `n` | `int` | **Required** | The total number of payments. |
| `P` | `float` | **Required** | The amount of a single payment. |
| `i` | `float` | **Required** | The effective interest rate per period (e.g., `0.08` for 8%). |
| `m_defer` | `int` | `0` | **Deferral periods**: How many periods pass *before* the annuity starts. |
| `m_post` | `int` | `0` | **Post-periods**: How many periods the capital grows *after* the last payment. |
| `eval_point`| `int` | `0` | Used only for `'current'` plot type. The specific time `t` where the equation of value is evaluated. |
| `annuity_type`| `str` | `'immediate'` | Payment timing: `'immediate'` (end of period) or `'due'` (beginning of period). |
| `plot_type` | `str` | `'pv'` | What to calculate/plot: `'pv'` (Present Value), `'fv'` (Future Value), or `'current'` (Equation of Value). |
